```python

new_data = test
new_data.reset_index()

sell_percent = 0.08
buy_percent =  0.04

for buy in range(20) :
    buy_percent = buy/100
    
    for sell in range(100) :
        sell_percent = sell/100
        new_high = []
        new_low = []
        new_status = []
        new_direction = []



        current_low = 1000000
        current_high = 10

        previous_price = new_data.Close[0]-10

        invest = 1000
        coin_amt = invest / new_data.Close[0]
        bank = 1000
        trade_amount = bank/10

        buy_permit = False
        data_range = len(new_data)



        for day in range(data_range) :
            close_price = new_data.Close[day]
            direction = int((close_price - current_low)/(current_high - current_low)*100)


            if direction<100 and direction >0 :
                if  previous_price > close_price :
                    status = "Down"
                else :
                    status = "Up"

            if close_price >= current_high :
                current_high = int(close_price)
                status="New High"
                moving_sell = (1-sell_percent) * current_high

            elif close_price <= current_low :
                current_low = int(close_price)
                status="New Low"
                moving_buy = (1+buy_percent) * current_low
                buy_permit = True


            if (status == "Down") and ((close_price/moving_sell) < 1) :
                status="Sell"
                if (coin_amt - (trade_amount/close_price)) >= 0 :
                    # print("== sell ==")
                    moving_sell = (1 - sell_percent) * moving_sell
                    coin_amt = coin_amt - (trade_amount/close_price)
                    bank = bank + trade_amount


            if (status == "Up") and ((close_price/moving_buy) > 1) :
                status="Buy"
                if (bank - trade_amount ) >= 0 and buy_permit :
                    # print("== buy ==")
                    moving_buy = buy_percent * moving_buy
                    coin_amt = coin_amt + (trade_amount/close_price)
                    bank = bank - trade_amount      
                    buy_permit = False




            new_high.append(current_high)
            new_low.append(current_low)
            new_status.append(status)
            new_direction.append(direction)
            previous_price = close_price
            if day == 749 :
                print(buy +1, sell+1, int(close_price), current_low, current_high, status, direction, int(coin_amt * close_price), int(bank), int(coin_amt * close_price + bank))

#new_data
    
```
